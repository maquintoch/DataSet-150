<h3>Defining the categorical and continuous values on the dataset.</h3>

```python
# Defyning the categorical and continuos variables.
Categorical_col = ['Country code', 'Country name', 'Series code', 'Series name']
Continuous_col = ['1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000',
                '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011']
```

```python
# Transforming all the continuous values -as we defined on the list above 'Continuous_col'- to numerical datatype.
for col in Continuous_col:
    data[col]=pd.to_numeric(data[col], errors='coerce')
    data[col]=np.nan_to_num(data[col])

# Transforming all the object values -as we defined on the list above 'Categorical_col'- to category datatype.
for col in Categorical_col:
    data[col] = data[col].astype('category')
```

```python
# I will plot the rate and year of the climate change in Argentina based on the topic of the changes. 

name_country.columns
coun_topic = name_country['Topic'].unique()
coun_topic

# I will select colors for each category on column Topic
colors = [plt.cm.tab10(i/float(len(coun_topic)-1)) for i in range(len(coun_topic))]
colors

plt.figure(figsize=(16, 10), dpi=80, facecolor="w", edgecolor="k")
for i, cat in enumerate(coun_topic):
    plt.scatter("Year", "Rate", data=name_country.loc[name_country.Topic == cat, :],
               s = 20, c=colors[i], label=str(cat))
plt.gca().set(xlabel='Year', ylabel='Rate')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title("Topic of clima change in Argentina based on year and rate")
plt.legend(fontsize=12)
plt.show()
```

<h4>I am creating my own function called "mean_of_rate" for computing and returning the mean of the column "Rate" on the dataframe "name_country" that contains all the values related to Argentina (Country name). By doing this I am adding a new column on that dataframe with the mean of the rate by year.</h4>

```python
def mean_of_rate(name_country):
    name_country['Mean']= name_country['Rate'].mean()
    return name_country

plot_mean_per_year =name_country.groupby('Year').apply(mean_of_rate)
plot_mean_per_year
    
```
